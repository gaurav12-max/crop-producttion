```python
plt.figure(figsize=(25,5))
ax=sns.countplot(data=crop1,x="Season")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize=(25,5))
ax=sns.countplot(data=crop1,x="Crop_Year")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
gb=crop1.groupby("State_Name").agg({"Crop_Year":"mean","Area":"mean","Production":"mean"})
print(gb)
```

```python
plt.figure(figsize=(50,100))
sns.heatmap(gb,annot=True)
plt.show()
```

```python
Yearly_Sales =  crop1[['State_Name', 'District_Name', 'Crop_Year', 'Season', 'Crop', 'Area',
       'Production']]
```

```python
plt.figure(figsize=(60,60))
Yearly_Sales01 = Yearly_Sales.groupby('Crop_Year').sum().reset_index()
sns.catplot(y = 'Production', x = 'Crop_Year', data = Yearly_Sales01, palette='Reds',kind="bar")
plt.xlabel('Year')
plt.ylabel('Production')
plt.title('Yearly Sales')
Yearly_Sales01[['Crop_Year', 'Production']]
```
